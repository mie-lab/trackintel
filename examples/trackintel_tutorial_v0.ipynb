# trackintel tutorial

### This tutorial will guide you through the diferent modules of trackintel and introduce you to the options to process your data

To import the framework run:

```python
import trackintel as ti
```

To verify the correct installation of the trackintel framework, test the installed version:

```python
ti.print_version()
```

The basis of evry process is the raw data consisting of coordinate tuples with timestamps, so called positionfixes. The raw data can be imported as a positionfixes object by using the read_positionfixes_csv method. Depending on your system preferences you may have to manually adapt the seperator of the csv.

```python
pfs = ti.io.read_positionfixes_csv('data/google_trajectory.csv',sep=';')
```

The imported positionfixes stored in pfs can now be plotted using the accessor as_positionfixes. This accessor will permit you to use the methods created for the positionfixes. The parameter plot_osm activates/deactivates the background map.

```python
pfs.as_positionfixes.plot(plot_osm=True)
```
